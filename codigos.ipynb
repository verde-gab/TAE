## 2. Modificação e Sanitização dos Dados
Aplicação de tratamentos textuais, tipagem de colunas de metadados temporais e categóricas.
O dataframe passará a ter colunas em 'lowercase'.

```r
# Contadores globais
total_asfactor <- 0
total_asdate   <- 0

remover_acentos <- function(x) {
  if (is.character(x) || is.factor(x)) {
    return(iconv(as.character(x), from = "UTF-8", to = "ASCII//TRANSLIT"))
  }
  return(x)
}

limpar_texto <- function(x, col) {
  if (col %in% c("ano", "mes")) {
    x <- trimws(as.character(x))
    x <- as.factor(x)
    total_asfactor <<- total_asfactor + 1
    return(x)
  }
  
  if (col == "ano_mes") {
    x <- as.Date(x)
    total_asdate <<- total_asdate + 1
    return(x)
  }
  
  if (is.character(x) || is.factor(x)) {
    x <- remover_acentos(x)
    x <- trimws(x)
    x <- gsub(" ", "_", x)
    x <- tolower(x)
    x <- as.factor(x)
    total_asfactor <<- total_asfactor + 1
  }
  return(x)
}

limpar_colunas <- function(df) {
  colnames(df) <- sapply(colnames(df), function(col) {
    col <- iconv(col, from = "UTF-8", to = "ASCII//TRANSLIT")
    col <- trimws(col)
    col <- gsub(" ", "_", col)
    col <- tolower(col)
    return(col)
  })
  return(df)
}

limpar_dados <- function(df) {
  df[] <- lapply(names(df), function(col) {
    limpar_texto(df[[col]], col)
  })
  return(df)
}

limpar_dataframe <- function(df) {
  df <- limpar_colunas(df)
  df <- limpar_dados(df)
  return(df)
}

# Execução da limpeza primária
df <- limpar_dataframe(df)

cat("Total convertido para factor:", total_asfactor, "\n")
cat("Total convertido para date:  ", total_asdate,   "\n")

# Tratamento Estrutural de Zeros (Filtro por Natureza da Carga)
tratar_zeros_teu <- function(df) {
  df$total_teu <- ifelse(
    df$total_teu == 0 & df$natureza_carga != "carga_conteinerizada",
    NA,
    df$total_teu
  )
  return(df)
}

tratar_zeros_unid <- function(df) {
  df$total_unid <- ifelse(
    df$total_unid == 0 &
      df$natureza_carga %in% c("granel_liquido", "granel_solido"),
    NA,
    df$total_unid
  )
  return(df)
}

# Aplica substituições estruturais
df <- tratar_zeros_unid(df)
df <- tratar_zeros_teu(df)

# Inspeciona estado final da tipagem
str(df)
```

## 3. Relatório Analítico
Geração do mini-relatório do dataset em formato limpo. Produz os arrays dimensionais das colunas numéricas (Média, Variância, Desvio Padrão) e fatores (Distribuição Frequencial).

```r
report_data <- function(df) {
  
  resultado_numero <- data.frame(
    coluna       = character(), min = numeric(), q1 = numeric(), mediana = numeric(),
    media        = numeric(), q3 = numeric(), max = numeric(), variancia = numeric(),
    desvio_padrao = numeric(), stringsAsFactors = FALSE
  )
  
  resultado_factor <- data.frame(
    coluna = character(), categoria = character(), frequencia = numeric(), stringsAsFactors = FALSE
  )
  
  cat("ESTRUTURA DO DATASET\n")
  cat("Linhas:", nrow(df), " | Colunas:", ncol(df), "\n\n")
  
  cat("RESUMO ESTATISTICO - NUMERICAS\n")
  for (col in names(df)) {
    x <- df[[col]]
    if (is.numeric(x)) {
      cal       <- summary(x, na.rm = TRUE)
      variancia <- var(x, na.rm = TRUE)
      desvio    <- sd(x,  na.rm = TRUE)
      resultado_numero <- rbind(
        resultado_numero,
        data.frame(
          coluna = col, min = as.numeric(cal["Min."]), q1 = as.numeric(cal["1st Qu."]),
          mediana = as.numeric(cal["Median"]), media = as.numeric(cal["Mean"]),
          q3 = as.numeric(cal["3rd Qu."]), max = as.numeric(cal["Max."]),
          variancia = variancia, desvio_padrao = desvio, stringsAsFactors = FALSE
        )
      )
    }
  }
  print(resultado_numero)
  cat(rep("=", 40), "\n")
  
  cat("\nFREQUENCIA DE CATEGORIAS (FACTOR)\n")
  for (col in names(df)) {
    x <- df[[col]]
    if (is.factor(x)) {
      cat(rep("=", 40), "\n")
      cat("COLUNA:", col, "\n")
      cat(rep("-", 40), "\n")
      freq <- head(summary(x), 10)
      
      temp <- data.frame(
        categoria = names(freq), frequencia = as.numeric(freq), stringsAsFactors = FALSE
      )
      print(temp)
      
      resultado_factor <- rbind(
        resultado_factor,
        data.frame(
          coluna = col, categoria = names(freq), frequencia = as.numeric(freq), stringsAsFactors = FALSE
        )
      )
    }
  }
  cat(rep("=", 40), "\n")
  
  return(list(numerico = resultado_numero, factor = resultado_factor))
}

relatorio <- report_data(df)
```

## 4. Visualizações e Projeções Gráficas
Análise visual interativa de Boxplots (discrepâncias), Histograma + Densidade de Pareto (variáveis contínuas) e Barplots categóricos.

```r
graficos <- function(df) {
  
  # 1. COLUNAS NUMÉRICAS — BOXPLOT
  for (col in names(df)) {
    x <- df[[col]]
    if (is.numeric(x) || is.integer(x)) {
      cat(rep("=", 40), "\n")
      cat("BOXPLOT:", col, "\n")
      par(mfrow = c(1, 1), mar = c(3, 3, 2, 1), oma = c(0, 0, 0, 0))
      boxplot(x, outline = FALSE, main = paste("Distribuicao de:", col),
              ylab = col, col = "lightblue", border = "darkblue", notch = FALSE)
    }
  }
  
  # 2. COLUNAS NUMERICAS — BARPLOT / HISTOGRAMA
  cols_num <- names(df)[sapply(df, function(x) is.numeric(x) || is.integer(x))]
  for (col in cols_num) {
    x <- df[[col]]
    x <- x[!is.na(x)]
    cat(rep("=", 40), "\n")
    cat("HISTOGRAMA:", col, "\n")
    par(mfrow = c(1, 1), mar = c(5, 4, 4, 2), oma = c(0, 0, 0, 0))
    hist(x, breaks = 50, freq = FALSE, main = paste("Histograma de:", col),
         xlab = col, ylab = "Densidade", col = "lightblue", border = "darkblue")
    lines(density(x), col = "red", lwd = 2)
    abline(v = mean(x), col = "darkgreen", lwd = 2, lty = 2)
    abline(v = median(x), col = "orange", lwd = 2, lty = 2)
    legend("topright", legend = c("Densidade", "Media", "Mediana"),
           col = c("red", "darkgreen", "orange"), lwd = 2, lty = c(1, 2, 2), cex = 0.8)
  }
  
  # 3. COLUNAS FACTOR — BARPLOT
  cols_factor <- names(df)[sapply(df, is.factor)]
  for (col in cols_factor) {
    x <- df[[col]]
    cat(rep("=", 40), "\n")
    cat("BARPLOT:", col, "\n")
    freq     <- sort(table(as.character(x)), decreasing = TRUE)
    freq_top <- head(freq, 10)
    
    par(mfrow = c(1, 1), mar = c(12, 4, 4, 2), oma = c(0, 0, 0, 0))
    barplot(freq_top, main = paste("Top 10:", col), ylab = "Frequencia",
            col = "lightblue", border = "darkblue", las = 2, cex.names = 0.8)
  }
  
  # 4. COLUNA DATE — SÉRIE TEMPORAL
  for (col in names(df)) {
    x <- df[[col]]
    if (inherits(x, "Date")) {
      cat(rep("=", 40), "\n")
      cat("SERIE TEMPORAL:", col, "\n")
      par(mfrow = c(1, 1), mar = c(5, 4, 4, 2), oma = c(0, 0, 0, 0))
      freq  <- table(x)
      datas <- as.Date(names(freq))
      plot(datas, as.numeric(freq), type = "l", main = paste("Serie temporal de:", col),
           xlab = "Data", ylab = "Frequencia", col = "darkblue", lwd = 2)
      lines(lowess(datas, as.numeric(freq)), col = "red", lwd = 2)
    }
  }
  return(df)
}

graficos_renderizados <- graficos(df)
```

## 5. Extração de Correlações e Inferência Estatística (Correlacao.R)
Cálculo analítico iterativo entre matrizes combinadas:
1. `Factor x Factor`: Extração de força direcional via Chi-Quadradro ($X^2$) e Coeficiente V de Cramér.
2. `Numeric x Numeric`: Determinação de monotonicidade através do Coeficiente de Spearman $\rho$ (robusto a caudas longas).
3. `Factor x Numeric`: Teste ANOVA não-paramétrico de Kruskal-Wallis com extração log-linear $R^2$.

```r
correlacionar <- function(df) {
  
  resultado_factor  <- data.frame(
    col_a = character(), col_b = character(), chi2 = numeric(), p_valor = numeric(),
    v_cramer = numeric(), conclusao = character(), stringsAsFactors = FALSE
  )
  resultado_numeric <- data.frame(
    col_a = character(), col_b = character(), spearman = numeric(), p_valor = numeric(),
    conclusao = character(), stringsAsFactors = FALSE
  )
  resultado_misto   <- data.frame(
    col_factor = character(), col_numeric = character(), kruskal_h = numeric(),
    p_valor = numeric(), r2 = numeric(), conclusao = character(), stringsAsFactors = FALSE
  )
  
  dados_num <- df
  for (col in names(dados_num)) {
    if (inherits(dados_num[[col]], "Date")) {
      dados_num[[col]] <- as.numeric(dados_num[[col]])
    }
  }
  
  cols_factor  <- names(dados_num)[sapply(dados_num, is.factor)]
  cols_numeric <- names(dados_num)[sapply(dados_num, is.numeric)]
  
  cols_numeric <- cols_numeric[!cols_numeric %in% c("numero_viagem", "metrica")]
  cols_factor  <- cols_factor[!cols_factor   %in% c("ano")]
  
  # ── CASO 1: FACTOR × FACTOR
  if (length(cols_factor) >= 2) {
    pares_factor <- combn(cols_factor, 2, simplify = FALSE)
    for (par in pares_factor) {
      col_a <- par[1]
      col_b <- par[2]
      tabela <- table(dados_num[[col_a]], dados_num[[col_b]])
      
      if (nrow(tabela) < 2 || ncol(tabela) < 2) next
      
      teste <- suppressWarnings(chisq.test(tabela))
      n        <- sum(tabela)
      chi2     <- as.numeric(teste$statistic)
      k        <- min(nrow(tabela), ncol(tabela)) - 1
      v_cramer <- sqrt(chi2 / (n * k))
      p_valor  <- teste$p.value
      
      if (p_valor < 0.05) {
        forca     <- ifelse(v_cramer >= 0.3, "forte", "fraca")
        conclusao <- paste("Associacao", forca, "- rejeita H0")
      } else {
        conclusao <- "Sem associacao - nao rejeita H0"
      }
      resultado_factor <- rbind(resultado_factor, data.frame(
        col_a = col_a, col_b = col_b, chi2 = chi2, p_valor = p_valor,
        v_cramer = round(v_cramer, 4), conclusao = conclusao, stringsAsFactors = FALSE
      ))
    }
  }

valid <- !is.na(x) & !is.na(y)
x <- x[valid]; y <- y[valid]
teste <- suppressWarnings(cor.test(x, y, method = "spearman"))
    
  # ── CASO 2: NUMERIC × NUMERIC
  if (length(cols_numeric) >= 2) {
    pares_numeric <- combn(cols_numeric, 2, simplify = FALSE)
    for (par in pares_numeric) {
      col_a <- par[1]
      col_b <- par[2]
      x <- dados_num[[col_a]]
      y <- dados_num[[col_b]]
      teste <- suppressWarnings(cor.test(x, y, method = "spearman", use = "complete.obs"))
      spearman <- as.numeric(teste$estimate)
      p_valor  <- teste$p.value
      
      if (p_valor < 0.05) {
        forca <- ifelse(abs(spearman) >= 0.6, "moderada/forte",
                        ifelse(abs(spearman) >= 0.3, "fraca", "muito fraca"))
        direcao  <- ifelse(spearman > 0, "positiva", "negativa")
        conclusao <- paste("Correlacao", forca, direcao)
      } else {
        conclusao <- "Sem correlacao - nao rejeita H0"
      }
      resultado_numeric <- rbind(resultado_numeric, data.frame(
        col_a = col_a, col_b = col_b, spearman = round(spearman, 4),
        p_valor = round(p_valor, 4), conclusao = conclusao, stringsAsFactors = FALSE
      ))
    }
  }
  
  # ── CASO 3: FACTOR × NUMERIC
  for (col_f in cols_factor) {
    for (col_n in cols_numeric) {
      x <- dados_num[[col_f]]
      y <- dados_num[[col_n]]
      validos <- !is.na(x) & !is.na(y)
      x <- x[validos]
      y <- y[validos]
      
      niveis_presentes <- unique(x)
      grupos_validos   <- tapply(y, x, function(g) length(unique(g)) > 1)
      n_grupos_validos <- sum(grupos_validos, na.rm = TRUE)
      
      if (length(niveis_presentes) < 2 || n_grupos_validos < 2) next
      
      teste     <- suppressWarnings(kruskal.test(y ~ x))
      p_valor   <- teste$p.value
      kruskal_h <- as.numeric(teste$statistic)
      
      var_total  <- var(y, na.rm = TRUE)
      vars_grupo <- tapply(y, x, var, na.rm = TRUE)
      ns_grupo   <- tapply(y, x, length)
      var_media  <- sum(ns_grupo * vars_grupo, na.rm = TRUE) / sum(ns_grupo)
      r2 <- 1 - (var_media / var_total)
      
      if (p_valor < 0.05) {
        forca     <- ifelse(r2 >= 0.3, "forte", "fraca")
        conclusao <- paste("Efeito", forca, "- rejeita H0 | R2:", round(r2, 4))
      } else {
        conclusao <- "Sem efeito - nao rejeita H0"
      }
      resultado_misto <- rbind(resultado_misto, data.frame(
        col_factor = col_f, col_numeric = col_n, kruskal_h = round(kruskal_h, 4),
        p_valor = round(p_valor, 4), r2 = round(r2, 4), conclusao = conclusao, stringsAsFactors = FALSE
      ))
    }
  }
  
  # ── MAPAS DE CALOR 
  if (length(cols_numeric) >= 2) {
    mat_cor <- cor(dados_num[, cols_numeric], method = "spearman", use = "complete.obs")
    n <- ncol(mat_cor)
    par(mfrow = c(1, 1), mar = c(12, 12, 4, 2), oma = c(0, 0, 0, 0))
    diag(mat_cor) <- NA
    image(1:n, 1:n, t(mat_cor[n:1, ]), col = colorRampPalette(c("darkblue", "white", "darkred"))(100),
          xaxt = "n", yaxt = "n", main = "Mapa de Calor - Correlacao Spearman", xlab = "", ylab = "")
    diag(mat_cor) <- 1
    axis(1, at = 1:n, labels = colnames(mat_cor), las = 2, cex.axis = 0.9, tick = FALSE)
    axis(2, at = 1:n, labels = rev(rownames(mat_cor)), las = 2, cex.axis = 0.9, tick = FALSE)
    for (i in 1:n) {
      for (j in 1:n) {
        text(i, n - j + 1, round(mat_cor[j, i], 2), cex = 0.9, col = "black")
      }
    }
  }
  
  if (length(cols_factor) >= 2) {
    n_f        <- length(cols_factor)
    mat_cramer <- matrix(NA, n_f, n_f, dimnames = list(cols_factor, cols_factor))
    for (i in seq_along(cols_factor)) {
      for (j in seq_along(cols_factor)) {
        if (i == j) {
          mat_cramer[i, j] <- 1
        } else {
          tab  <- table(dados_num[[cols_factor[i]]], dados_num[[cols_factor[j]]])
          if (nrow(tab) < 2 || ncol(tab) < 2) next
          chi2 <- as.numeric(suppressWarnings(chisq.test(tab))$statistic)
          n    <- sum(tab)
          k    <- min(nrow(tab), ncol(tab)) - 1
          mat_cramer[i, j] <- sqrt(chi2 / (n * k))
        }
      }
    }
    par(mar = c(8, 8, 4, 2))
    image(1:n_f, 1:n_f, t(mat_cramer), col = colorRampPalette(c("white", "darkblue"))(100),
          xaxt = "n", yaxt = "n", main = "Mapa de Calor - V de Cramer (Factor x Factor)", xlab = "", ylab = "")
    axis(1, at = 1:n_f, labels = cols_factor, las = 2, cex.axis = 0.7)
    axis(2, at = 1:n_f, labels = cols_factor, las = 2, cex.axis = 0.7)
    for (i in 1:n_f) {
      for (j in 1:n_f) {
        text(j, i, round(mat_cramer[i, j], 2), cex = 0.7)
      }
    }
  }
  
  return(list(factor = resultado_factor, numeric = resultado_numeric, misto = resultado_misto))
}

resultados_correlacao <- correlacionar(df)

# Exibição analítica dos testes
print("Resultados de Spearman (Numérico x Numérico):")
print(resultados_correlacao$numeric)

print("Resultados V de Cramér (Fator x Fator):")
print(resultados_correlacao$factor)

print("Resultados de Kruskal-Wallis + R² (Fator x Numérico):")
print(resultados_correlacao$misto)
```
