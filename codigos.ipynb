# Dados de movimentação de cargas do Porto de Santos.
## Apresentação da tabela de variáveis.

| Variável | Tipo em R | Descrição | Valores |
| :-------------- | :-------: | :------- | :------- |
| TOTAL_TEU | Integer | Unidade de medida padrão para contêineres de 20 pés. | > 0 |
| NUMERO_VIAGEM | Integer | Identificador numérico de uma escala portuária, atribuído pelo sistema de origem. | > 0 |
| TIPO_NAVEGACAO | Character | Classifica o tipo de navegação logística. | Longo Curso / Cabotagem |
| MOVIMENTO | Character | Caracteriza o tipo de movimentação realizada pelo navio. | Convencional / Remoção / Transbordo |
| ANO | Integer | Identifica o ano de determinado registro. | [2005-2026] |
| TERMINAIS | Character | Classifica o terminal de operção que foi realizado o processo logístico. | Texto Livre |
| TOTAL_TONELADAS | Double | Determina a quantidade em peso de determinada carga | > 0 |
| NATUREZA_CARGA  | Character | Classifica a natureza física da carga movimentada. | Texto Livre |
| TOTAL_UNID      | Integer   | Quantidade de unidades físicas movimentadas na operação. | > 0 |
| MES             | Integer   | Exibe o mês resgistrado daquel movimentação. | [01-12] |
| CLASSENAVIO     | Character | identifica o tipo de navio que realizou a movimentação. | Texto Livre |
| SENTIDO         | Character | Determina qual o processo que a carga teve no porto. | Embarque / Desembarque |
| TIPO_INSTALACAO | Character | Classifica o tipo de instalação portuária onde ocorreu a movimentação. | Texto Livre |
| BERCOS          | Character | Identifica o berço de atracação utilizado na operação. | Texto Livre |
| MERCADORIAS     | Character | Classifica exatamente a carga transportada. | Texto Livre |
| METRICA         | Double    | Coluna redundante de TOTAL_TONELADAS. | > 0 |
| ANO_MES         | Character | Exibe a notação universal para data conforme ISO 8601.  | AAAA-MM-DD |
